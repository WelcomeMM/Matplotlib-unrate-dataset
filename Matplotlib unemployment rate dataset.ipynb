Comparaison between 1948 - 1949

```python
help(plt.Figure.add_subplot)
```

```python
first_year = unrate[0:12]
second_year = unrate[12:24]

fig = plt.figure(figsize=(15,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(first_year["DATE"], first_year["UNRATE"])
ax2.plot(second_year["DATE"], second_year["UNRATE"])

ax1.title.set_text("Unemployment rate in 1948")
ax2.set_title("Unemployment rate in 1949")

plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5, 1, i+1)
    first_index = i * 12
    last_index = (i + 1) * 12
    subset = unrate[first_index : last_index]
    ax.plot(subset["DATE"], subset["UNRATE"])
    
plt.show()
```

```python
unrate["MONTH"] = unrate["DATE"].dt.month

fig = plt.figure(figsize=(6,3))

plt.plot(unrate[0:12]["MONTH"], unrate[:12]["UNRATE"], c="red")
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["UNRATE"], c="blue")

plt.show()
```

```python
help(plt.title)
```

```python
fig = plt.figure(figsize=(10,6))

colors = ["red", "blue", "green", "orange", "black"]

for i in range(5):
    
    start_index = i * 12
    end_index = (i+1) * 12
    year = 1948 + i
    
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"], subset["UNRATE"], c=colors[i], label=year)


plt.legend(loc="upper left")
plt.title("Unemployment rate from 1948 to 1952")
plt.xlabel("Month (from 1st to 12th)")
plt.ylabel("Unemployment rate (%)")


plt.show()
```
